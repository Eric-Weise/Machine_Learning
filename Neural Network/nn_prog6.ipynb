```python
from numpy import loadtxt

def load_image_file(inputFile):

    '''
    load image file and return X and y
    The image file has the following format for each line:
        label, pix-1, pix-2, pix-3, ...pix-784
        
    Returns:
    X: should be [m, 784], where m is the number of images
    y: size should be [m, 1]
    '''
    #Load the dataset
    data = loadtxt(inputFile, delimiter=',')

    # build X (descriptive feature data) and y (target data)
    X = data[:, 1:785]
    y = data[:, 0]
    
    #number of training samples
    return X, y

X_train, y_train = load_image_file('mnist_train_6000.csv')
X_test, y_test = load_image_file('mnist_test_1000.csv')
print('Rows: %d, columns: %d' % (X_train.shape[0], X_train.shape[1]))
#print(X_train[0])

print(y_train.shape)
```

```python
def onehot(y, n_classes):
        """Encode labels into one-hot representation
        """
        onehot = np.zeros((y.shape[0], n_classes))
        for idx, val in enumerate(y.astype(int)):
            onehot[idx,val] = 1
        return onehot.T
    
# testing
print((y_test[:5]))
print(onehot(y_test[:5], 10))
```

```python
def initialize_parameters(n_x, n_h, n_y):
    """
    Argument:
    n_x -- size of the input layer
    n_h -- size of the hidden layer
    n_y -- size of the output layer
    
    Returns:
    parameters -- python dictionary containing your parameters:
                    W1 -- weight matrix of shape (n_h, n_x)
                    b1 -- bias vector of shape (n_h, 1)
                    W2 -- weight matrix of shape (n_y, n_h)
                    b2 -- bias vector of shape (n_y, 1)
    """
    
    np.random.seed(1)
    
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros(shape=(n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros(shape=(n_y, 1))
    
    assert(W1.shape == (n_h, n_x))
    assert(b1.shape == (n_h, 1))
    assert(W2.shape == (n_y, n_h))
    assert(b2.shape == (n_y, 1))
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```
